```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/HeshamTay/Datasets/main/Countries.csv')
```

```python
df.drop('R&D', axis=1, inplace=True)
```

```python
for col in df.columns:
  if df[col].dtype == float:
    df[col].fillna(df[col].mean(), inplace=True)
```

```python
#Let's check if any of the countries doesn't have values in at least one year (In order to be able to compare several countries on the same graph)
l = []
for cont in df['Country Name'].unique():
  if df[df['Country Name'] == cont]['Year'].nunique() != 23:
    l.append()
l
```

Great! All listed countries have entries on each of the 23 years
